## Блок функций

```python
def pre_research(data, column, bins_count, range_min=99, range_max=99):
    """Функция предварительного анализа данных
    Данная функция выведет уникальные значения указанного
    столбца и построит гистограмму.

    Для использования функции необходимо передать ей
    следующие аргументы:
    
    data - массив данных, которые нужно анализировать;
    column - название столбца, по которому проводится анализ;
    bins_count - количество корзин в гистограмме;
    range_min, range_max - минимальное и максимальное
        граничные значения, выводимые на гистограмму
        по умолчанию задаются минимумом и максимумом значений столбца, 
        при необходимости можно внести нужные занчения для 
        масштабирования гистограммы.
    """
    
    print(f'Уникальные значения {column}: \n', data[column].sort_values().unique())
    print(f'1-й персентиль {data[column].quantile(.1)}')
    print(f'99-й персентиль {data[column].quantile(.99)}')
    # В столбце есть выбросы значений, построим гистограмму для того, 
    # чтобы понять как часто встречаются большие и очень малые значения
    fig = plt.figure(figsize=(15, 6)) 
    if range_min==range_max:
        range_min = data[column].min()
        range_max = data[column].max()
    data[column].hist(bins=bins_count,range=(range_min,range_max))
    plt.title(f"{column} histogram")
    plt.ylabel(f"{column} count")
    plt.xlabel(column)

    plt.show()
    
def find_id (data, column):
    """Функция выделения id из строки по символу "_"

    Для использования функции необходимо передать ей
    следующие аргументы:
    
    data - массив исходных данных;
    column - столбец, в котором необходимо провести изменения - 'название_столбца';
    new_column - название столбца, которым необходимо заменить столбец 'id' - 'название_столбца'.
    
    """
    # Используем регулярные выражения для выделения номеров звонков 
    data[column] = data[column].astype('str').apply(lambda x: re.sub(r'\w+\_', '', x)).astype('int')

```

## Расчёты и добавление результатов в таблицу

Объединим таблицу users с таблицами internet, messages, calls по значению столбца 'user_id'


```python
calls = calls.merge(users ,how='inner', on='user_id')
internet = internet.merge(users ,how='inner', on='user_id')
messages = messages.merge(users ,how='inner', on='user_id')
```

Создадим столбец с месяцами в соответствии с датой

```python
calls['month'] = calls['call_date'].apply(lambda x: x.month)
messages['month'] = messages['message_date'].apply(lambda x: x.month)
internet['month'] = internet['session_date'].apply(lambda x: x.month)
```

Выделим только удавшиеся звонки для подсчета количества

```python
calls_sucsess = calls.query('duration != 0')
```

Сформируем сводные таблицы с подсчетом метрик для каждого типа данных о деятельности пользователя

```python
calls_pivot = calls_sucsess.pivot_table(index=['user_id','month'], 
                                values = 'duration', 
                                aggfunc = ('count','sum')).reset_index()
calls_pivot = calls_pivot.rename(columns={'count':'call_count','sum':'sum_minut'})
```

```python
messages_pivot = messages.pivot_table(index=['user_id','month'], 
                                   values = 'id', 
                                   aggfunc = 'count').reset_index()
messages_pivot = messages_pivot.rename(columns={'id':'msg_count'})
```

```python
internet_pivot = internet.pivot_table(index=['user_id','month'], 
                                   values = 'mb_used', 
                                   aggfunc = 'sum').reset_index()
```

Сформируем общую таблицу путем объединения всех таблиц с привязкой к 'user_id','month'

```python
pivot_all_data = calls_pivot.copy()
```

```python
pivot_all_data = pivot_all_data.merge(messages_pivot, on=['user_id','month'],how='outer')
```

```python
pivot_all_data = pivot_all_data.merge(internet_pivot, on=['user_id','month'],how='outer')
```

```python
pivot_all_data = pivot_all_data.merge(users, on='user_id')
pivot_all_data = pivot_all_data.rename(columns = {'tariff':'tariff_name'})
```

```python
pivot_all_data = pivot_all_data.merge(tariffs, on='tariff_name')
```

Рассчитаем на сколько каждый пользователь вышел за пределы лимита использования трафика

```python
pivot_all_data['dif_gb'] = (pivot_all_data['mb_used'] - pivot_all_data['mb_per_month_included'])/1000
pivot_all_data['dif_gb'] = pivot_all_data['dif_gb'].fillna(0)
pivot_all_data['dif_gb'] = pivot_all_data['dif_gb'].apply(lambda x: math.ceil(x))
pivot_all_data.loc[pivot_all_data['dif_gb']<0,'dif_gb'] = 0
```

```python
pivot_all_data['dif_minutes'] = pivot_all_data['sum_minut'] - pivot_all_data['minutes_included']
pivot_all_data['dif_minutes'] = pivot_all_data['dif_minutes'].fillna(0)
pivot_all_data.loc[pivot_all_data['dif_minutes']<0,'dif_minutes'] = 0
```

```python
pivot_all_data['dif_msg'] = pivot_all_data['msg_count'] - pivot_all_data['messages_included']
pivot_all_data['dif_msg'] = pivot_all_data['dif_msg'].fillna(0)
pivot_all_data.loc[pivot_all_data['dif_msg']<0,'dif_msg'] = 0
```

```python
pivot_all_data['income'] = pivot_all_data['dif_msg']*pivot_all_data['rub_per_message'] + \
                            pivot_all_data['dif_minutes']*pivot_all_data['rub_per_minute'] + \
                            pivot_all_data['dif_gb']*pivot_all_data['rub_per_gb'] + \
                            pivot_all_data['rub_monthly_fee']
pivot_all_data['income'] = pivot_all_data['income'].astype('int')
```

В сводной таблице есть пропуски в значениях call_count sum_minut и msg_count, скорее всего эти пропуски появились естественным путем - абоненты не используют звонки и смс, заполним пропуски нулями

```python
pivot_all_data['call_count'] = pivot_all_data['call_count'].fillna(0)
pivot_all_data['sum_minut'] = pivot_all_data['sum_minut'].fillna(0)
pivot_all_data['msg_count'] = pivot_all_data['msg_count'].fillna(0)
```

Уберем лишние денные из сводной таблицы

```python
pivot_all_data.info()
```

```python
pivot_all_data = pivot_all_data.drop(axis = 1, 
                                     labels = ['churn_date',
                                              'age',
                                              'first_name',
                                              'last_name',
                                              'reg_date',])
```

```python
mean_by_month = pivot_all_data.pivot_table(index = ['tariff_name','month'], 
                           values = ['income',
                                     'mb_used',
                                     'sum_minut',
                                     'msg_count',
                                     'dif_gb',
                                     'dif_minutes',
                                     'dif_msg',
                                    ], 
                           aggfunc = 'mean')
mean_by_month
```

Построим графики для каждого столбца с группировкой по тарифу

```python
mean_by_month = mean_by_month.reset_index()
```

```python
list_column = ['income',
               'mb_used',
               'sum_minut',
               'msg_count',
              ]
titles_of_figure = ['Средний заработок на тарифе за месяц',
                    'Срейднее количество использованных мегабайт трафика за месяц',
                    'Средняя продолжительность звонков за месяц',
                    'Среднее количество СМС сообщений за месяц',
                   ]
zip_titles = zip(list_column, titles_of_figure)
dict_of_coll = dict(zip_titles)

count_lines = ['smart', 'ultra']

for column in list_column:
    fig = plt.figure(figsize=(15, 6)) 
    for i in count_lines:
        plot_data = mean_by_month[mean_by_month['tariff_name']==i]
        plt.plot(plot_data['month'], 
                 plot_data[column], 
                 linestyle= '-', marker='o',
                 linewidth=1, 
                 alpha=0.9,
                )
    ax = fig.add_subplot(1, 1, 1)
    major_ticks = np.arange(1, plot_data['month'].max()+1, 1)
    ax.set_xticks(major_ticks)
    ax.grid(which='both')
    ax.grid(which='major', alpha=0.5)
    plt.grid(True)
    plt.title(dict_of_coll[column])
    plt.legend(count_lines)
    plt.show()
    
```

Проверим следующие гипотезы:

1. средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются;
2. средняя выручка пользователи из Москвы отличается от выручки пользователей из других регионов.


1. Формируем нолувую гипотегу H0: "Гипотеза о равенстве средней выручки с тарифов Ultra и Smart"

```python
alpha = 0.05

results = st.ttest_ind(stat_research.loc[stat_research['tariff_name']=='smart', 'income'],
                       stat_research.loc[stat_research['tariff_name']=='ultra', 'income'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Это означает, что мы не можем принять гипотезу о равенстве средней выручки тарифов Ultra и Smart, скорее всего выручка с тарифов отличается

2.  Формируем нолувую гипотегу H0: "Средняя выручка компании от пользователей из Москвы равна выручке с пользователей из остальных регионов"

```python
alpha = 0.05

results = st.ttest_ind(stat_research.loc[stat_research['city']=='Москва', 'income'],
                       stat_research.loc[stat_research['city']!='Москва', 'income'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Можем сделать вывод, что не можем утвержать о том, что пользователи из Москвы приносят оператору выручку, отличную от пользователей из регионов
